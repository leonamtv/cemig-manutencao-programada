```python
import pandas as pd
```

```python
file_path = '../downloads/downloaded_sheet_1650500151.xls'
```

```python
df = pd.read_excel(file_path)
new_header = df.iloc[1]
df = df[2:] 
df.columns = new_header 
df.head(30)
```

```python
df_timoteo = df.loc[df['Cidade'] == 'TIMOTEO']
df_timoteo.head(30)
```

```python
from datetime import datetime

COLUMN_INI = 'INÍCIO'
COLUMN_FIM = 'FIM'
```

```python
df_filtered = df.loc[df['Cidade'] == 'TIMOTEO'].copy()
df_filtered[COLUMN_INI] = pd.to_datetime(df_filtered[COLUMN_INI])
df_filtered[COLUMN_FIM] = pd.to_datetime(df_filtered[COLUMN_FIM])
df_filtered[COLUMN_INI+'DT'] = df_filtered[COLUMN_INI].dt.date
df_filtered[COLUMN_FIM+'DT'] = df_filtered[COLUMN_FIM].dt.date
df_date = df_filtered.loc[df_filtered[COLUMN_INI+'DT'] <= datetime(2022, 4, 24).date()].loc[datetime(2022, 4, 24).date() <= df_filtered[COLUMN_FIM+'DT']]
df_date.head(10)
```
